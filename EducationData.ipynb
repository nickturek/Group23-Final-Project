## **Education**
Education is the cornerstone of a healthy and thriving community. When people are educated they learn about the importance of hard work, and can make more informed decisions about leadership, both of which contribute to a more productive community. Let's look at a dataset that compares degree level and income among total population living in each neighborhood in Pittsburgh.

```python
import pandas as pd
import numpy as np
%matplotlib inline
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("https://data.wprdc.org/dataset/40188e1c-6d2e-4f20-9391-607bd3054949/resource/f7b19c6c-aa66-419b-b0e1-9998d7ddfcbc/download/education-income.csv")
df.head()
```

This dataset has a lot of information, but let's focus it to just the percent of people in each neighborhood that have any type of college degree. Recieving a degree from a higher-level institution can mean more job opportunities, better pay, and a better overall community, so a neighborhood with the more college-educated residents will be a better neighborhood.

```python
df.set_index('Neighborhood', inplace = True)
bachelors = df["Edu. Attainment: Bachelor's Degree (2010)"].str.rstrip('%').astype('float')
assoc = df["Edu. Attainment: Assoc./Prof. Degree (2010)"].str.rstrip('%').astype('float')
postgrad = df["Edu. Attainment: Postgraduate Degree (2010)"].str.rstrip('%').astype('float')
total_college = bachelors + assoc + postgrad
print(total_college)
```

Great! But now we need to order it from highest percentage to lowest.

```python
sorted_df = total_college.sort_values(ascending=False)
sorted_df
```

Looking good! Now lets graph it.

```python
sorted_df.plot(kind='bar')
```

Whoa, that's a lot to look at! Let's narrow it to just the top ten with the highest percentage of residents with college degrees.

```python
sorted_df.head(10).plot(kind='bar')
```
